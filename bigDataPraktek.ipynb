```python
diabetes = pd.read_csv("diabetes.csv")
diabetes.head(10)
```

```python
diabetes.isnull().sum()
```

```python
diabetes.hist(column='Age', bins=50);
```

```python
hitung = diabetes['Insulin']
modus = hitung.mode()
median = hitung.median()
mean = hitung.mean()
range = hitung.max() - hitung.min()
standardeviasi = hitung.std()
print('Modusnya adalah : ',modus[0])
print('Meannya adalah : ',mean)
print('Mediannya adalah : ',median)
print('Standar deviasinya adalah: ',standardeviasi)
print('Range-nya adalah: ',range)
```
